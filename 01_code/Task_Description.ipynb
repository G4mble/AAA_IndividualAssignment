# How to design successful broker strategies in a competitive simulation framework?

Power TAC is a competitive simulation that models a “liberalized” retail electrical energy market, where competing business entities or “brokers” offer energy services to customers through tariff contracts, and must then serve those customers by trading in a wholesale market. 
The retail market is a tariff market, in which customers are able to choose among tariff contract offerings from the competing brokers. Customers are models of household, business, and institutional users of electric power, as well as small-scale producers of power that own solar arrays or small wind turbines (Ketter et al. 2013, 2016a, 2016b).
Once a year a Power TAC competition is launched in which loose groups and institutions from all over the world are granted the option to implement a broker and test their strategies within a competitive simulation framework. 
More precisely, brokers are challenged to maximize their profits by buying and selling energy in the wholesale and retail markets, subject to fixed costs and constraints; the winner of an individual “game” is the broker with the highest bank balance at the end of a simulation run. 
Costs include fees for publication and withdrawal of tariffs, and distribution fees for transporting energy to their contracted customers. 
Costs are also incurred whenever there is an imbalance between a broker’s total contracted energy supply and demand within a given time slot.
However, the awards ceremony after the finals of each competition is only a starting point for the participating parties to gain insights into the underlying drivers of their performance.
Therefore, detailed log files are provided afterwards. 
Thereby, participants can reproduce single rounds and trace individual target figures and actions such as tariff transactions during the course of competitions.
Imagine you are a business entity who participated within the Power TAC finals of either 2016, 2017 or 2018. 
We provide you the respective log files comprising all the information you need to understand what drove your final performance. 
As you are planning to win next year’s competition (again), you are interested in deriving recommended actions by applying data analytics.

### Which brokers do you tend to outperform and which not?

```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt

brokerSpecificData = pd.read_csv("../00_data/finals_2018_games.csv", sep=";", decimal=".") # read input data
brokerSpecificData = brokerSpecificData[["gameName", "AgentUDE", "Bunnie", "COLDPower18", "CrocodileAgent", "EWIIS3", "SPOT", "VidyutVanika"]]
brokerSpecificData = brokerSpecificData.dropna(subset=["Bunnie"]) # drop all NaN for specific broker, i.e. only consider the games in which Bunnie participates
competingBrokers = brokerSpecificData.columns[1:] # derive list of all brokers
competingBrokers = competingBrokers.drop(["Bunnie"]) # derive list of competing brokers except Bunnie

numberOfGamesPlayedByBunnie = brokerSpecificData["Bunnie"].count()

# derive number of games played by Bunnie with each broker separately
numberOfGamesPlayedWithEachCompetingBroker = {}
for broker in competingBrokers:
    tempData = brokerSpecificData[["SPOT", "Bunnie"]]
    tempData = tempData.dropna()
    numberOfGamesPlayedWithEachCompetingBroker[broker] = tempData["Bunnie"].count()

for broker in competingBrokers:
    brokerSpecificData[broker] = np.where(pd.isna(brokerSpecificData[broker]) , sys.maxsize, brokerSpecificData[broker]) # replace the NaN with a large number to ensure that a broker cannot be outperformed in a game he did not participate in
    brokerSpecificData[broker] = np.where(brokerSpecificData["Bunnie"] > brokerSpecificData[broker], 1, 0) # 1, if Bunnie outperformed the respective broker in the current game; 0, otherwise

# count how often Bunnie outperforms each broker
outperformCount = brokerSpecificData.select_dtypes(pd.np.number).sum().drop(["Bunnie"]).astype('int64')

# print results

print("\n\n")
print("The following graph shows the number of times each broker is outperformed by the broker Bunnie.")
print("NOTE: Games in which a broker did not participate are counted as 'not outperformed by Bunnie'.")

print("\n\n")
print("The broker Bunnie did participate in a total of " + str(numberOfGamesPlayedByBunnie) + " games.")

for broker in outperformCount.index:
    print("\n")
    print(broker + " and Bunnie participated in " + str(numberOfGamesPlayedWithEachCompetingBroker[broker]) + " games together.")
    print(broker + " is outperformed " + str(outperformCount[broker]) + " times by the broker Bunnie.")
    print("This corresponds to " + "{:.2f}".format((outperformCount[broker] / numberOfGamesPlayedWithEachCompetingBroker[broker]) * 100) + "% of the total number of games played involving Bunnie and " + broker + ".")

x = outperformCount.index
y = outperformCount.values
mean = outperformCount.values.sum() / competingBrokers.size

print("\n")
print("Minimum (red): " + str(min(y)))
print("Mean (yellow): " + str(mean))
print("Maximum (green): " + str(max(y)))

print("\n")
print("The broker VidyutVanika and AgentUDE tend to be outperformed least often by the broker Bunnie. They are outperformed by Bunnie " + str(outperformCount["VidyutVanika"]) + " times and " + str(outperformCount["AgentUDE"]) + " times respectively.")
print("\n")
print("The broker EWIIS3 tends to be outperformed by the boker Bunnie most often. He is outperformed " + str(outperformCount["EWIIS3"]) + " times by Bunnie.")

print("\n")
plt.bar(x, y)
plt.axhline(max(y), linestyle="--", color="g")
plt.axhline(min(y), linestyle="--", color="r")
plt.axhline(mean, linestyle="--", color="y")
plt.xlabel("Agent Name")
plt.xticks(rotation=90)
plt.ylabel("# of Times Outperformed by Bunnie")
plt.show()
```

### Cash position
##### What is the relation between the cash position and all other information provided? 

For a specific "Hour of Day" (denoted by "t"), the "Broker Net Cash Position" in t is the sum of the following columns in t:
    
    Tariff Transaction -- Credit
    Tariff Transaction -- Debit	
    Tariff Transaction -- Customer Credit	
    Tariff Transaction -- Customer Debit	
    Market Transaction -- Credit	
    Market Transaction -- Debit	
    Balancing Transaction -- Credit	
    Balancing Transaction -- Debit	
    Distribution Transaction -- Credit	
    Distribution Transaction -- Debit	
    Capacity Transaction -- Credit	
    Capacity Transaction -- Debit	
    Balancing Control Event -- Credit	
    Balancing Control Event -- Debit	
    Bank Transaction -- Credit	
    Bank Transaction -- Debit

plus/minus the following column in t-1:
    Broker Net Cash Position
    
So in essence, the Broker Net Cash Position in t consists of the Broker Net Cash position in t-1 plus all transaction credits and debits in t.

##### Information and business goal
Your business goal is to design competitive tariffs.
In which way is the individual information relevant in order to reach that goal?

The "Distribution Transaction -- Credit/Debit" are not really relevant to tariff design, since they represent a fixed fee which is only variable in the amount (kWh) of electricity distributed.

The "Balancing Transaction -- Credit/Debit" and the "Balancing Control Event -- Credit/Debit" data are only relevant for tariff design when the broker decides to issue tariffs for which he is unable to provide the full capacity and hence the discrepancy between offered and provided energy is settled via the balancing market.

The "Capacity Transaction -- Credit/Debit" information is relevant for tariff design, since it influences the decision of a broker to offer tariffs that provide more/less energy during peak demand hours. It might also influence dynamic prices of tariffs for different consumption periods, e.g. higher prices during peak demand hours due to capacity fees.

The "Market Transaction" information is not relevant for tariff design, since it only represents clearing price * amount traded.

The "Tariff Transaction" data is highly relevant for tariff design, since these information give insight into when new tariffs are published and what the fees for publishing are. Furthermore, information is provided about the amount of money received from a customer at which timeslot which might in turn be a valuable information source for deciding on dynamic prices for the respective tariffs.

Due to the different distribution of gameSize = 7 and gameSize = 4, we choose three "medium performance" games with gameSize = 7 and one "medium performance" game with gameSize = 4.

The final selection looks like this:
- gameSize = 7: 1 x "winner"; 1 x "worst performer"; 3 x "medium performance"
- gameSize = 4: 1 x "winner"; 1 x "worst performer"; 1 x "medium performance"
- gameSize = 2: 1 x "winner"; 1 x "worst performer"; 0 x "medium performance"

!Please see below for the explicit gameNames of the selected games!

```python
import pandas as pd
import numpy as np
import sys

raw_data = pd.read_csv("../00_data/finals_2018_games.csv", sep=";", decimal=".")
competingBrokers = ["AgentUDE", "COLDPower18", "CrocodileAgent", "EWIIS3", "SPOT", "VidyutVanika"]

# determine games won by Bunnie [0 || 1]
# replace the NaN with sys.maxsize * -1 to ensure that a broker does not prevent others from winning a game he did not participate in
tempData_win = raw_data.copy(deep=True)
for broker in competingBrokers:
    tempData_win[broker] = np.where(pd.isna(tempData_win[broker]) , (sys.maxsize * -1), tempData_win[broker])
    
# indicate a game won by bunnie with 1; 0 otherwise
raw_data["game_won_by_bunnie"] = np.where(((tempData_win["AgentUDE"] < tempData_win["Bunnie"]) & (tempData_win["COLDPower18"] < tempData_win["Bunnie"]) & (tempData_win["CrocodileAgent"] < tempData_win["Bunnie"]) & (tempData_win["EWIIS3"] < tempData_win["Bunnie"]) & (tempData_win["SPOT"] < tempData_win["Bunnie"]) & (tempData_win["VidyutVanika"] < tempData_win["Bunnie"])) , 1, 0)

# replace the NaN with sys.maxsize to ensure that a broker does not prevent others from losing a game he did not participate in
tempData_lose = raw_data.copy(deep=True)
for broker in competingBrokers:
    tempData_lose[broker] = np.where(pd.isna(tempData_lose[broker]) , sys.maxsize, tempData_lose[broker])
    
# indicate a game where bunnie was the worst performer by 1; 0 otherwise
raw_data["game_worst_performer_bunnie"] = np.where(((tempData_lose["AgentUDE"] > tempData_lose["Bunnie"]) & (tempData_lose["COLDPower18"] > tempData_lose["Bunnie"]) & (tempData_lose["CrocodileAgent"] > tempData_lose["Bunnie"]) & (tempData_lose["EWIIS3"] > tempData_lose["Bunnie"]) & (tempData_lose["SPOT"] > tempData_lose["Bunnie"]) & (tempData_lose["VidyutVanika"] > tempData_lose["Bunnie"])) , 1, 0)

# split data by games won by Bunnie
raw_gamesWon = raw_data[(raw_data.game_won_by_bunnie == 1)]

# split data by games where Bunnie was the worst performer
raw_gamesWorstPerformer = raw_data[(raw_data.game_worst_performer_bunnie == 1)]

# split data by games NOT won AND where Bunnie was NOT the worst performer
raw_gamesMediumPerformance = raw_data[((raw_data.game_worst_performer_bunnie == 0) & (raw_data.game_won_by_bunnie == 0))]
# drop all NaN for specific broker, i.e. only consider the games in which Bunnie participates
raw_gamesMediumPerformance = raw_gamesMediumPerformance.dropna(subset=["Bunnie"])

# split data by gamesSize
gamesWon_gameSize_7 = raw_gamesWon[(raw_gamesWon.gameSize == 7)]
gamesWon_gameSize_4 = raw_gamesWon[(raw_gamesWon.gameSize == 4)]
gamesWon_gameSize_2 = raw_gamesWon[(raw_gamesWon.gameSize == 2)]

worstPerformer_gameSize_7 = raw_gamesWorstPerformer[(raw_gamesWorstPerformer.gameSize == 7)]
worstPerformer_gameSize_4 = raw_gamesWorstPerformer[(raw_gamesWorstPerformer.gameSize == 4)]
worstPerformer_gameSize_2 = raw_gamesWorstPerformer[(raw_gamesWorstPerformer.gameSize == 2)]

mediumPerformance_gameSize_7 = raw_gamesMediumPerformance[(raw_gamesMediumPerformance.gameSize == 7)]
mediumPerformance_gameSize_4 = raw_gamesMediumPerformance[(raw_gamesMediumPerformance.gameSize == 4)]

# select final games
selection_gameWon_gameSize_7 = gamesWon_gameSize_7.iloc[0]
selection_gameWon_gameSize_4 = gamesWon_gameSize_4.iloc[0]
selection_gameWon_gameSize_2 = gamesWon_gameSize_2.iloc[0]

selection_worstPerformer_gameSize_7 = worstPerformer_gameSize_7.iloc[0]
selection_worstPerformer_gameSize_4 = worstPerformer_gameSize_4.iloc[0]
selection_worstPerformer_gameSize_2 = worstPerformer_gameSize_2.iloc[0]

selection_mediumPerformer_gameSize_7_threeTimes = mediumPerformance_gameSize_7.iloc[:3]
selection_mediumPerformer_gameSize_7_threeTimes = selection_mediumPerformer_gameSize_7_threeTimes.set_index("gameId", drop=True)
selection_mediumPerformer_gameSize_4 = mediumPerformance_gameSize_4.iloc[0]

print("For gameSize = 7 we selected the following games:")
print("\t'winner':\t\t" + selection_gameWon_gameSize_7["gameName"])
print("\t'worst performer':\t" + selection_worstPerformer_gameSize_7["gameName"])
print("\t'medium performance (x3)':")
for item in selection_mediumPerformer_gameSize_7_threeTimes.gameName:
    print("\t\t\t\t" + item)

print("\n")
print("For gameSize = 4 we selected the following games:")
print("\t'winner':\t\t" + selection_gameWon_gameSize_4["gameName"])
print("\t'worst performer':\t" + selection_worstPerformer_gameSize_4["gameName"])
print("\t'medium performance':\t" + selection_mediumPerformer_gameSize_4["gameName"])

print("\n")
print("For gameSize = 2 we selected the following games:")
print("\t'winner':\t\t" + selection_gameWon_gameSize_2["gameName"])
print("\t'worst performer':\t" + selection_worstPerformer_gameSize_2["gameName"])
```
